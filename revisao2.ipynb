```python
import pandas as pd
```

```python
df = pd.read_csv('vendas_vestuario_revisao.csv')
df
```

```python
df['Valor Vendido'] = df['quantidade']*df['valor_unitario'] #cria coluna na tabela 
df
```

```python
df_valor_vendido_vendedor = df.groupby('vendedor')['Valor Vendido'].sum().reset_index()  #grupando dados do vendedor por valor vendido
df_valor_vendido_vendedor
```
