```python
import pandas as pd
import numpy as np
```

```python
dados_1sem = 'Preços semestrais - AUTOMOTIVOS_2025.01.csv'
df_1sem = pd.read_csv(dados_1sem, sep=';')
df_1sem.head(3)
```

```python
dados_2sem = 'Preços semestrais - AUTOMOTIVOS_2025.02.csv'
df_2sem = pd.read_csv(dados_2sem, sep=';')
#df_2sem = [coluna1.colunax, colunaz] - escolhemos colunas
df_2sem.head(3)
```

```python
#concatenar arquivos
df_anp = pd.concat([df_1sem,df_2sem], ignore_index = True)   #concatenou 1o sem e 2o sem; ignora index para ignorar indices do 1o e 2o sem
```

```python
#transformar números que estão com virgula/string para ponto/número para permitir cálculos posteriores
#após mudança para float, se rodar novamente, vai dar erro pois o número já é float e não string
df_anp['Valor de Venda'] = df_anp['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
df_gasolina = df_anp.loc[df_anp['Produto'] == 'GASOLINA']
df_etanol = df_anp.loc[df_anp['Produto'] == 'ETANOL']
```

```python
#agrupar produto por estado para depois fazer os cálculos
df_media_est_gaso = df_gasolina.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()  #grupando valor de venda calculando média

df_media_est_gaso = df_media_est_gaso.sort_values(by='Valor de Venda', ascending = False)  #grupando valor de venda calculando média
df_media_est_gaso

```

```python
df_media_est_gaso.head()
```

```python
#agrupar produto por estado para depois fazer os cálculos
df_media_est_etanol = df_etanol.groupby('Estado - Sigla')['Valor de Venda'].mean().reset_index()  #grupando valor de venda calculando média

df_media_est_etanol = df_media_est_etanol.sort_values(by='Valor de Venda', ascending = False)  #grupando valor de venda calculando média
df_media_est_etanol
```

```python
df_media_est_etanol.head()
```

```python
#comparação regional de preços gasolina
df_media_regiao_gasolina = df_gasolina.groupby('Regiao - Sigla')['Valor de Venda'].mean().reset_index()  #grupando valor de venda calculando média
df_media_regiao_gasolina
```

```python
#comparação regional de preços etano
df_media_regiao_etanol = df_etanol.groupby('Regiao - Sigla')['Valor de Venda'].mean().reset_index()  #grupando valor de venda calculando média
df_media_regiao_etanol
```
